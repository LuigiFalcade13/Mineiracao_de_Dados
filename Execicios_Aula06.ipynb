```
#2
import numpy as np

array1 = np.random.randint(0, 51, size=10)
array2 = np.random.randint(0, 51, size=10)

soma = array1 + array2
subtracao = array1 - array2
multiplicacao = array1 * array2

divisao = np.divide(array1.astype(float), array2.astype(float), where=array2 != 0)
divisao[array2 == 0] = np.nan  # Substituir NaN nos casos onde o divisor era zero

print("Array 1:", array1)
print("Array 2:", array2)
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

```
#9
import pandas as pd
import numpy as np

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']

np.random.seed(0)
quantidade = np.random.randint(1, 21, size=10)
preco = np.random.uniform(10, 100, size=10).round(2)
produto = np.random.choice(produtos, size=10)

df_vendas = pd.DataFrame({
    'Produto': produto,
    'Quantidade': quantidade,
    'Preço': preco
})

df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

df_vendas_maior_que_10 = df_vendas[df_vendas['Quantidade'] > 10]

print("DataFrame de Vendas com Valor Total:")
print(df_vendas)
print("\nDataFrame de Vendas (Quantidade > 10) com Valor Total:")
print(df_vendas_maior_que_10)
```

```
#10
import pandas as pd
import numpy as np

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']

np.random.seed(0)
quantidade = np.random.randint(1, 21, size=10)
preco = np.random.uniform(10, 100, size=10).round(2)
produto = np.random.choice(produtos, size=10)

df_vendas = pd.DataFrame({
    'Produto': produto,
    'Quantidade': quantidade,
    'Preço': preco
})

df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

df_agrupado = df_vendas.groupby('Produto')['Quantidade'].sum().reset_index()

print("DataFrame Agrupado pelo Produto com Soma das Quantidades:")
print(df_agrupado)
```

```
#11
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
meses = np.arange(1, 13)
precos = np.random.uniform(10, 100, size=12).round(2)

df_preco = pd.DataFrame({
    'Mês': meses,
    'Preço': precos
})

plt.figure(figsize=(10, 6))
plt.plot(df_preco['Mês'], df_preco['Preço'], marker='o', linestyle='-', color='b')

plt.title('Variação de Preços de um Produto ao Longo de 12 Meses')
plt.xlabel('Mês')
plt.ylabel('Preço')
plt.xticks(df_preco['Mês'])
plt.grid(True)

plt.show()
```

```
#15
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = np.random.randint(1, 101, size=5)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

plt.figure(figsize=(10, 6))
bars = plt.bar(df_vendas['Produto'], df_vendas['Quantidade'], color='skyblue', label='Quantidade Vendida')

plt.title('Comparação de Quantidade Vendida de Produtos')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.legend()

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
#16
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = np.random.randint(1, 101, size=5)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Produto', y='Quantidade', data=df_vendas, palette='Blues_d')

plt.title('Comparação de Quantidade Vendida de Produtos')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```
#17
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(0)
n_pessoas = 100
altura = np.random.uniform(150, 200, size=n_pessoas)
peso = np.random.uniform(50, 100, size=n_pessoas)

df_dados = pd.DataFrame({
    'Altura': altura,
    'Peso': peso
})

plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df_dados, scatter_kws={'color': 'skyblue', 's': 100}, line_kws={'color': 'red'})

plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```
#19
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(0)
n_amostras = 100
variavel1 = np.random.rand(n_amostras)
variavel2 = np.random.rand(n_amostras) * 10
variavel3 = variavel1 * 2 + np.random.randn(n_amostras)
variavel4 = variavel2 / 2 + np.random.randn(n_amostras)

df_dados = pd.DataFrame({
    'Variável 1': variavel1,
    'Variável 2': variavel2,
    'Variável 3': variavel3,
    'Variável 4': variavel4
})

correlacao = df_dados.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Heatmap da Correlação entre Variáveis')

plt.show()
```

```
#20
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(0)
n_amostras = 100

variavel1 = np.random.normal(loc=50, scale=10, size=n_amostras)
variavel2 = np.random.normal(loc=30, scale=5, size=n_amostras)
variavel3 = variavel1 * 0.5 + np.random.normal(loc=0, scale=5, size=n_amostras)
variavel4 = variavel2 * 2 + np.random.normal(loc=0, scale=5, size=n_amostras)

df_dados = pd.DataFrame({
    'Variável 1': variavel1,
    'Variável 2': variavel2,
    'Variável 3': variavel3,
    'Variável 4': variavel4
})

sns.pairplot(df_dados)

plt.suptitle('Pairplot das Variáveis', y=1.02)

plt.show()
```
